2.- Comenzamos con el primer codigo que es el de **usuarios.py**: no olvides abrir una terminal solo para ejecutar este codigo en tu computadora con las instrucciones del paso 1 

```python
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/usuarios/<int:usuario_id>', methods=['GET'])
def obtener_usuario(usuario_id):
    # Lógica para obtener información del usuario desde una base de datos
    usuario = {
        'id': usuario_id,
        'nombre': 'Felix Mata',
        'correo': 'migfel@gmail.com'
    }
    return jsonify(usuario)

if __name__ == '__main__':
    app.run()

```

**3.- Ahora veamos el Microservicio de Productos.py:** 
 no olvides abrir una terminal solo para ejecutar este codigo en tu computadora con las instrucciones del paso 1 y usando el puerto 8081 y el nombre productos


```python
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/productos/<int:producto_id>', methods=['GET'])
def obtener_producto(producto_id):
    # Lógica para obtener información del producto desde una base de datos
    producto = {
        'id': producto_id,
        'nombre': 'iPhone 12',
        'precio': 999.99
    }
    return jsonify(producto)

if __name__ == '__main__':
    app.run()
```
